```python
### TextLoader
# from langchain_core.document_loaders import TextLoader as TL # might throw errors
from langchain_community.document_loaders import TextLoader as TL2 # alternative way to import TextLoader.

loader = TL2("../data/text_files/python-desc.txt", encoding="utf-8")
document = loader.load()

print(document)
```

```python
# DirectoryLoader
from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader(
    "../data/text_files", 
    glob="*.txt", # pattern to match files in the directory.
    loader_cls=TL2, # the loader class to use for loading the files.
    loader_kwargs={"encoding": "utf-8"}, # additional arguments to pass to the loader class.
    show_progress=True, 
)

documents = dir_loader.load()
documents
```

```python
# DirectoryLoader
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader

dir_loader = DirectoryLoader(
    "../data/pdf", 
    glob="*.pdf", # pattern to match files in the directory.
    loader_cls=PyMuPDFLoader, # the loader class to use for loading the files.
    # loader_kwargs={"encoding": "utf-8"}, # additional arguments to pass to the loader class.
    show_progress=True, 
)

pdf_documents = dir_loader.load()
pdf_documents
```
